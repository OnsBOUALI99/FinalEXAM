```python
import matplotlib.pyplot as plt #we import matplotlib so we can use charts
```

```python
# Import the libraries
import pandas as pd
import matplotlib.pyplot as plt


# Filter the data: only keep rows where people said "yes" to Bank_deposit
yes_deposits = df[df['Bank_deposit'] == 'yes']

# Count how many people from each job said yes
job_yes = yes_deposits['job'].value_counts()

# Now we plot it
job_yes.plot(kind='bar')  # Make a bar chart

# Add a title and labels
plt.title("Number of YES Deposits by Job Type")
plt.xlabel("Job")
plt.ylabel("Count")
plt.xticks(rotation=45)  # Rotate job labels to make them readable
plt.tight_layout()       # Makes sure nothing is cut off

# Show the plot
plt.show()
```

```python
# Count how many people are in each education category
edu_counts = df['education'].value_counts()

# Plot pie chart
edu_counts.plot(kind='pie')
plt.title("Education Levels of Clients")
plt.ylabel("")  # Hide default label
plt.tight_layout()
plt.show()
```

```python
# Plot a histogram to see age distribution
df['age'].plot(kind='hist', bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of People")
plt.tight_layout()
plt.show()
```

```python
# Filter only 'yes' responses and count marital status
marital_yes = df[df['Bank_deposit'] == 'yes']['marital'].value_counts()

# Plot bar chart
marital_yes.plot(kind='bar')
plt.title("Positive Bank deposits by Marital Status")
plt.xlabel("Marital Status")
plt.ylabel("total")
plt.tight_layout()
plt.show()
```

```python
# Chart 5: How many people already have a personal loan

import matplotlib.pyplot as plt

# Count how many people have a loan vs not
loan_counts = df['loan'].value_counts()

# Make a bar chart
loan_counts.plot(kind='bar')

# Add title and labels
plt.title("Clients With and Without Loans")
plt.xlabel("Has Personal Loan")
plt.ylabel("Number of Clients")
plt.tight_layout()

# Show the chart
plt.show()
```

```python
#Interpretation:
#In this analysis, I looked at five different charts to understand the clients better.
#Most clients in the dataset do not have personal loans, which means the bank is reaching people who might still be open to new offers.
#When looking at who said yes to opening a deposit, married people were the most likely to agree, followed by divorced and single clients.
#The age chart showed that most clients are between 50 and 60 years old, with very few younger or much older people.
#In terms of education, most clients completed secondary school, while fewer had primary or tertiary education.
#Finally, the job chart showed that retired people said yes the most, followed by people in management and technician roles.
#These charts suggest that older, married, and retired clients are the most responsive to the bank's deposit offers.
```
